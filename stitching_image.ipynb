## Stitching image

```python
# importing required modules
import cv2 as cv 
from matplotlib  import pyplot as plt 
%matplotlib inline
```

```python
# Read image
img = cv.imread("landscape.jpg")

#Verify read image
plt.imshow(img)
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
```

```python
#Create sub images from the main image

#create sub image along width
(ht,wd) = img.shape[:2]
x = 0
y = 0
wd_half = wd/2

sub1 = img[y:y+int(ht), x:x+int(wd_half)]
plt.imshow(sub1)
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
```

```python
wd_half = wd/2

sub2 = img[y:y+int(ht), x+int(wd_half):x+wd]

plt.imshow(sub2)
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
```

```python
# stitch sub1 and sub2 to get full image
import numpy as np

stitch1 = np.concatenate((sub1, sub2), axis=1)

plt.imshow(stitch1)
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
```

```python
# stitching together to get overlapping image
wd_brk = wd/2 + 200
sub3 = img[y:y+int(ht), x:x+int(wd_brk)]


plt.imshow(sub3)
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
```

```python
stitch2 = np.concatenate((sub3, sub2), axis=1)

plt.imshow(stitch2)
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
```
